# svm.SVC()
- epsilon (float, default=0.1)

```python
import numpy as np
import pylab as pl

X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
pl.scatter(X, y, c='k', label='data')
```

```python
from sklearn.svm import SVR
```

```python
svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1, epsilon = 0.1)
y_rbf = svr_rbf.fit(X, y).predict(X)

pl.scatter(X, y, c='k', label='data')
pl.plot(X, y_rbf, c='g', label='RBF model')
```

```python
svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1, epsilon = 0.0001)
y_rbf = svr_rbf.fit(X, y).predict(X)

pl.scatter(X, y, c='k', label='data')
pl.plot(X, y_rbf, c='g', label='RBF model')
```

```python
svr_poly = SVR(kernel='poly', C=1e4, degree=3, epsilon = 0.001)
y_poly  = svr_poly.fit(X, y).predict(X)

pl.scatter(X, y, c='k', label='data')
pl.plot(X, y_poly , c='g', label='poly model')
```

```python
svr_lin = SVR(kernel='linear', C=1e4)
y_lin = svr_lin.fit(X, y).predict(X)

pl.scatter(X, y, c='k', label='data')
pl.plot(X, y_lin , c='g', label='lineae model')
```
